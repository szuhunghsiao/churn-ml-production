### Identify categorical columns

```python
obj_col = df.select_dtypes(include=['object']).columns.tolist()
num_col = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(obj_col)
print(num_col)
```

```python
# Get each Object Column's unique data and number of unique of data
categorical_col = []
numerical_col = []
for col in obj_col:
    print(f" - {col} \t: {df[col].nunique()} \t, {df[col].unique()}")
    if df[col].nunique() < 10:
        categorical_col.append(col)
    elif df[col].nunique() < df.shape[0]:
        numerical_col.append(col)
print(f"\n\nCategorical Columns: {categorical_col}")
print(f"\n\nNumerical Columns: {numerical_col}")
```

```python
for col in num_col:
    print(f" - {col} \t: {df[col].nunique()} \t, {df[col].unique()}")
    if df[col].nunique() < 10:
        categorical_col.append(col)
    elif df[col].nunique() < df.shape[0]:
        numerical_col.append(col)
print(f"\n\nCategorical Columns: {categorical_col}")
print(f"\n\nNumerical Columns: {numerical_col}")
```

```python
for col in categorical_col:
    print(f"\n - {col} \t:")
    for category in df[col].unique():
        print(f"\t - {category}: {sum(df[col]==category)}, percentage: {round(sum(df[col]==category)/df.shape[0],2)}")
```
